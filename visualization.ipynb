Мозаичные диаграммы

```python
# переимнование столбца
df = df.rename(columns = {'Кто вы сейчас?' : 'activity', 'Как вы считаете, у вас есть вредные привычки?' : 'bad_habits'})
```

```python
N = 3

#считаем колисество школьников без вредных привычек и с вредными привычками
schoolboy0 = df[(df.activity == 'Школьник')][(df.bad_habits == 'Нет')]['А сколько на дополнительное обучение?'].count()
schoolboy1 = df[(df.activity == 'Школьник')][(df.bad_habits == 'Да')]['А сколько на дополнительное обучение?'].count()

#считаем колисество студентов без вредных привычек и с вредными привычками
student0 = df[(df.activity == 'Студент')][(df.bad_habits == 'Нет')]['А сколько на дополнительное обучение?'].count()
student1 = df[(df.activity == 'Студент')][(df.bad_habits == 'Да')]['А сколько на дополнительное обучение?'].count()

#считаем колисество рабочих без вредных привычек и с вредными привычками
working0 = df[(df.activity == 'Рабочий')][(df.bad_habits == 'Нет')]['А сколько на дополнительное обучение?'].count()
working1 = df[(df.activity == 'Рабочий')][(df.bad_habits == 'Да')]['А сколько на дополнительное обучение?'].count()


yes = (schoolboy1, student1, working1)
no = (schoolboy0, student0, working0)


ind = np.arange(N)    
width = 0.35

p1 = plt.bar(ind, yes, width)
p2 = plt.bar(ind, no, width,
             bottom=yes)

plt.xticks(ind, ('Школьник', 'Студент', 'Рабочий'))
plt.legend((p1[0], p2[0]), ('Нет', 'Да'))

plt.show()
```

```python
N = 3

# считаем колисество школьников без вредных привычек и с вредными привычками
schoolboy0 = df[(df.activity == 'Школьник')][(df.bad_habits == 'Нет')]['А сколько на дополнительное обучение?'].count()
schoolboy1 = df[(df.activity == 'Школьник')][(df.bad_habits == 'Да')]['А сколько на дополнительное обучение?'].count()

# считаем колисество студентов без вредных привычек и с вредными привычками
student0 = df[(df.activity == 'Студент')][(df.bad_habits == 'Нет')]['А сколько на дополнительное обучение?'].count()
student1 = df[(df.activity == 'Студент')][(df.bad_habits == 'Да')]['А сколько на дополнительное обучение?'].count()

# считаем колисество рабочих без вредных привычек и с вредными привычками
working0 = df[(df.activity == 'Рабочий')][(df.bad_habits == 'Нет')]['А сколько на дополнительное обучение?'].count()
working1 = df[(df.activity == 'Рабочий')][(df.bad_habits == 'Да')]['А сколько на дополнительное обучение?'].count()


# Нормализация
s_b = schoolboy0 + schoolboy1
schoolboy0 = schoolboy0/s_b
schoolboy1 = schoolboy1/s_b

w = working0 + working1
working0 = working0/w
working1 = working1/w

s = student0 + student1
student0 = student0/s
student1 = student1/s

yes = (schoolboy1, student1, working1)
no = (schoolboy0, student0, working0)


ind = np.arange(N)    
width = 0.35

p1 = plt.bar(ind, yes, width)
p2 = plt.bar(ind, no, width,
             bottom=yes)

plt.xticks(ind, ('Рабочий', 'Студент', 'Школьник'))
plt.legend((p1[0], p2[0]), ('Да', 'Нет'))

plt.show()
```

Диаграмма размаха

```python
game = pd.DataFrame()
game = pd.read_csv('high_diamond_ranked_10min.csv', sep = ',')
```

```python
game
```

```python
# записываем серии с количеством защитных тотемов, установленных победтившей и проигравшей командой 
loss = game[(game.blueWins == 0)]['blueWardsPlaced']
win = game[(game.blueWins == 1)]['blueWardsPlaced']

plt.boxplot([loss, win])

plt.show()
```

```python
# записываем серии с количеством золота в минуту победтившей и проигравшей командой 
loss = game[(game.blueWins == 0)]['blueGoldPerMin']
win = game[(game.blueWins == 1)]['blueGoldPerMin']

plt.boxplot([loss, win])

plt.show()
```

Графики плотности

```python
import seaborn as sns; sns.set(color_codes=True)

# Draw Plot
plt.figure(figsize=(16,10))
sns.kdeplot(game.loc[game['blueWins'] == 1, "blueGoldPerMin"], shade=True, color="g", label="win", alpha=.7)
sns.kdeplot(game.loc[game['blueWins'] == 0, "blueGoldPerMin"], shade=True, color="deeppink", label="loss", alpha=.7)


# Decoration
plt.title('GAME', fontsize=22)
plt.legend()
plt.show()
```
